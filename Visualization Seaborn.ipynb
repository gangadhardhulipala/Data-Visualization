```python
# Importing the necessary libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Importing the dat from csv file as a dataframe

movies = pd.read_csv("Movies.csv", encoding = "ISO-8859-1")
```

```python
movies.head()
```

```python
# Scatter plot view

sns.scatterplot(x = "Runtime", y="Critic.Score", data=movies)
plt.show()
```

```python
# Adding the categorical variable

sns.scatterplot(x = "Runtime", y="Critic.Score", data=movies, hue = "Awards")
plt.show()
```

```python
# Adding transperancy factor for better visualizations

sns.scatterplot(x = "Runtime", y="Critic.Score", data=movies, hue = "Awards", alpha = 0.4)
plt.show()
```

```python
# Controlling the colors for the categorical variable

sns.scatterplot(x = "Runtime", y="Critic.Score", data=movies, hue = "Awards", palette = {"Awards":"red", "No Awards":"blue","Won Awards":"green"})
plt.show()
```

```python
# Controlling the order of the legend for the categorical variable

sns.scatterplot(x = "Runtime", y="Critic.Score", 
                data=movies, hue_order = ["Won Awards", "No Awards"], 
                palette = {"Awards":"red", "No Awards":"blue","Won Awards":"green"}, alpha = 0.3)
plt.show()
```

```python
# Adding a count plot 

sns.countplot(x = "Awards", data = movies)
```

```python
# Adding one more dependecy for the counts - in this case its distribution

sns.countplot(x = "Awards", data = movies, hue = "Distribution")
```

```python
# Using a relational plot for simiplification

sns.relplot(x = "Runtime", y="Critic.Score", data=movies, kind = "scatter", hue = "Distribution")
```

```python
# Adding a categotical variable to the size 

sns.relplot(x = "Runtime", y="Critic.Score", data=movies, kind = "scatter", hue = "Distribution", size = "Rating")
```

```python
# Adding the transperacy field

sns.relplot(x = "Runtime", y="Critic.Score", data=movies, kind = "scatter", 
            hue = "Distribution", style = "Rating", alpha = 0.6)
```

```python
# Creating subplots

sns.relplot(x = "Runtime", y="Critic.Score", data=movies, kind = "scatter", 
            hue = "Distribution", alpha = 0.6, col = "Rating")
```

```python
# Creating subplots by controlling number of columns

sns.relplot(x = "Runtime", y="Critic.Score", data=movies, kind = "scatter", 
            hue = "Distribution", alpha = 0.6, col = "Rating", col_wrap = 2)
```

```python
# Creating subplots

sns.relplot(x = "Runtime", y="Critic.Score", data=movies, kind = "scatter", 
            hue = "Distribution", alpha = 0.6, col = "Rating", row = "Awards")
```

```python
# Adding line plot

sns.relplot(x="Critic.Score", y="Box.Office", 
            data=movies, kind="line")
```

```python
# Removing confidence intervals for the line plot

sns.relplot(x="Critic.Score", y="Box.Office", 
            data=movies, kind="line", ci = None)
```

```python
# Count plots using categoricl plots

sns.catplot(x = "Awards", data = movies, hue = "Distribution", kind = "count")
```

```python
# Bar plots

sns.catplot(x = "Awards", y = "Box.Office", data = movies, hue = "Distribution", kind = "bar", capsize = 0.05)
```

```python
# Changing the order of the categorical variable in bar plots

sns.catplot(x = "Awards", y = "Box.Office", data = movies, hue = "Distribution", kind = "bar", capsize = 0.05, order = ["Won Awards", "No Awards"])
```

```python
# Box plots

sns.catplot(x = "Awards", y = "Box.Office", data = movies, hue = "Distribution", kind = "box", order = ["Won Awards", "No Awards"])
```

```python
# Box plots - Omitting the outliers

sns.catplot(x = "Awards", y = "Box.Office", data = movies, hue = "Distribution", kind = "box", 
            order = ["Won Awards", "No Awards"], sym = "")
```

```python
# Box plots - Changing the interquartile range

sns.catplot(x = "Awards", y = "Box.Office", data = movies, hue = "Distribution", kind = "box", 
            order = ["Won Awards", "No Awards"], whis = 0.5)
```

```python
# Box plots : Custom range for the whiskers

sns.catplot(x = "Awards", y = "Box.Office", data = movies, hue = "Distribution", kind = "box", 
            order = ["Won Awards", "No Awards"], whis = [25, 75])
```

```python
# Box plots : With different visualizations

sns.catplot(x = "Awards", y = "Box.Office", data = movies, hue = "Distribution", kind = "boxen", 
            order = ["Won Awards", "No Awards"])
```

```python
# Importing the dat from csv file as a dataframe

movies = pd.read_csv("Movies.csv", encoding = "ISO-8859-1")
movies.head()
```

```python
# Point plots :
sns.catplot(x = "Year", y = "Box.Office", data = movies, hue = "Distribution", 
            kind = "point")
```

```python
# Point plots : vertical tick labels
sns.catplot(x = "Year", y = "Box.Office", data = movies, hue = "Distribution", 
            kind = "point", dodge = True, orient = "v")
plt.xticks(rotation = 90)
```

```python
# Point plots : with no join
sns.catplot(x = "Year", y = "Box.Office", data = movies, hue = "Distribution", 
            kind = "point", dodge = True, join = False, orient = "v")
plt.xticks(rotation = 90)
```

```python
from numpy import median

# Point plots : with no join
sns.catplot(x = "Year", y = "Box.Office", data = movies, hue = "Distribution", 
            kind = "point", dodge = True, join = False, estimator = median)

plt.xticks(rotation = 90)
```

```python
# Strip plots

sns.stripplot(x = "Rating", y="Box.Office", data = movies, jitter = True)
```

```python
# Swarm plots to visualize the overlapping points

sns.catplot(x = "Rating", y="Box.Office", data = movies, kind = 'swarm')
```

```python
# Swarm plots with points differentiated by distribution

sns.catplot(x = "Rating", y="Box.Office", hue = "Distribution", data = movies, kind = 'swarm')
```

```python
# Violin plots

sns.catplot(x = "Rating", y="Box.Office", hue = "Distribution", data = movies, kind = 'violin')
```

```python
# Violin plots split with categorical variable distribution

sns.catplot(x = "Rating", y="Box.Office", hue = "Distribution", data = movies, kind = 'violin', split = True)
```

```python
# Pair plots

x = sns.PairGrid(movies)
x = x.map(plt.scatter)
```

```python
# Controlling diagnol plots

x = sns.PairGrid(movies)
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
```

```python
# Adding a hue variable to the pair plot

x = sns.PairGrid(movies, hue = "Distribution", palette = "coolwarm")
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
x = x.add_legend()
```

```python
# Specifying the columns in the pair plot as well as using hue

x = sns.PairGrid(movies, hue = "Distribution", vars = ["Runtime", "Box.Office"])
x = x.map_diag(plt.hist)
x = x.map_offdiag(plt.scatter)
```

```python
# Instead of on and off diagnal we are controlling diagnal, upper and lower parts of pair plot

x = sns.PairGrid(movies, hue = "Distribution")
x = x.map_diag(plt.hist)
x = x.map_upper(plt.scatter)
x = x.map_lower(sns.kdeplot)
x = x.add_legend()
```
